```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("users.data",sep="\t",names=["user_id","item_id","rating","timestap"])
```

```python
data.head()
```

```python
len(data)
```

```python
movies=pd.read_csv("movie_id_titles.csv")
```

```python
movies
```

```python
data=pd.merge(data,movies,on="item_id")
```

```python
data[data["title"] == "Star Wars (1977)"]["rating"].mean()
```

```python
data.head()
```

```python
table = data.pivot_table(index="user_id",columns="title",values="rating")
table.head()
```

```python
starwars = table["Star Wars (1977)"]
starwars.head()
```

```python
type(starwars)
```

```python
correlation = table.corrwith(starwars)
```

```python
correlation
```

```python
correlation2 = pd.DataFrame(correlation,columns=["Correlation"])
correlation2.dropna(inplace=True)
correlation2.head()
```

```python
correlation2.sort_values("Correlation",ascending=False).head(15)
```

```python
data.head()
```

```python
data.drop((["timestap"]),axis=1)
```

```python
ratings=pd.DataFrame(data.groupby("title")["rating"].mean())
ratings.sort_values("rating",ascending=False).head()
```

```python
ratings["votecount"] = pd.DataFrame(data.groupby("title")["rating"].count())
```

```python
ratings.head()
```

```python
ratings.sort_values("votecount",ascending=False).head()
```

```python
correlation2.head()
```

```python
correlation2=correlation2.join(ratings["votecount"])
correlation2.head()
```

```python
correlation2[correlation2["votecount"] > 100].sort_values("Correlation",ascending=False).head()
```
